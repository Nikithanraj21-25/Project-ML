## **Dataset Loading and Removal of Unwanted Columns**

```python
dataset = pd.read_csv('carPricePredictionDataset.csv')
dataset = dataset.drop(['car_ID'], axis=1)
dataset
```

## **Finding Unique Values in Object Data Type Columns**

```python
print(dataset['fueltype'].unique())
print(dataset['aspiration'].unique())
print(dataset['doornumber'].unique())
print(dataset['carbody'].unique())
print(dataset['drivewheel'].unique())
print(dataset['enginelocation'].unique())
print(dataset['enginetype'].unique())
print(dataset['cylindernumber'].unique())
print(dataset['fuelsystem'].unique())
print(dataset['CarName'].unique())
```

## **Mapping of Unique Values to Numerical Values**

```python
dataset['fueltype'] = dataset['fueltype'].map({'diesel' : 0, 'gas' : 1})
dataset['aspiration'] = dataset['aspiration'].map({'std' : 0, 'turbo' : 1})
dataset['doornumber'] = dataset['doornumber'].map({'two' : 0, 'four' : 1})
dataset['carbody'] = dataset['carbody'].map({'convertible' : 1, 'hatchback' : 2, 'sedan' : 3, 'wagon' : 4, 'hardtop' : 5})
dataset['drivewheel'] = dataset['drivewheel'].map({'fwd' : 0, '4wd' : 1, 'rwd' : 2})
dataset['enginelocation'] = dataset['enginelocation'].map({'front' : 0, 'rear' : 1})
dataset['enginetype'] = dataset['enginetype'].map({'dohc' : 1, 'ohcv' : 2, 'ohc' : 3, 'l' : 4, 'rotor' : 5, 'ohcf' : 6, 'dohcv' : 7})
dataset['cylindernumber'] = dataset['cylindernumber'].map({'four' : 4, 'six' : 6, 'five' : 5, 'three' : 3, 'twelve' : 12, 'two' : 2, 'eight' : 8})
dataset['fuelsystem'] = dataset['fuelsystem'].map({'mpfi' : 1, '2bbl' : 2, 'mfi' : 3, '1bbl' : 4, 'spfi' : 5, '4bbl' : 6, 'idi' : 7, 'spdi' : 8})

```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the car names
dataset['CarName'] = label_encoder.fit_transform(dataset['CarName'])

```

## **Processed Dataset**

```python
dataset
```

## **Dataset Summarization**

```python
print(dataset.shape)
print(dataset.head(5))
```

## **Segregating Dataset into Input and Output**

```python
x = dataset.iloc[:, :-1].values
x
```

```python
y = dataset.iloc[:,-1].values
y
```

## **Splitting Dataset for Training and Testing**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
print(x_train.shape)
print(x_test.shape)
```

## **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

## **Model Training**

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

## **Prediction for Test Dataset**

```python
y_pred = model.predict(x_test)
```

## **Model Validation**

```python
from sklearn.metrics import r2_score,mean_squared_error

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("Root Mean Square Error : ", rmse)

r2score = r2_score(y_test, y_pred)
print("R2Score : ", r2score*100)
```
